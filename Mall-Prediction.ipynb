```python
df=pd.read_csv('/kaggle/input/customer-segmentation-tutorial-in-python/Mall_Customers.csv')
```

```python
print(df.shape)
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
```

```python
X=df.iloc[:,3:].values
X.shape
```

```python
X
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(X)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel('No. of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
km=KMeans(n_clusters=5)
```

```python
y_pred=km.fit_predict(X)
```

```python
y_pred
```

```python
X[y_pred==1,1]
```

```python
plt.scatter(X[y_pred==0,0],X[y_pred==0,1],color='red')
plt.scatter(X[y_pred==1,0],X[y_pred==1,1],color='black')
plt.scatter(X[y_pred==2,0],X[y_pred==2,1],color='green')
plt.scatter(X[y_pred==3,0],X[y_pred==3,1],color='orange')
plt.scatter(X[y_pred==4,0],X[y_pred==4,1],color='blue')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
```
